```python
#Interpolation Problem Statement
import matplotlib.pyplot as plt
import numpy as np
#These points form the basis for the quadratic equation y=x^2, but only shows the defined points of the equation, not the smooth curve of a quadratic.
x = [-5, -3, -1, 0, 1, 3, 5]
y = [25,9, 1, 0, 1, 9, 25]
plt.plot(x, y, 'o')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()

#Linear Interpolation
#Using the same points as above, the points are joined to create a closer approximation of the quadratic.
from scipy.interpolate import interp1d
f = interp1d(x, y)
x1 = np.arange(-5, 6)
y1 = f(x1)
plt.plot(x1, y1,'r')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()

#Cubic Spline Interpolation
#This creates a smooth curve, using our data to create y=x^2 exactly.
from scipy.interpolate import CubicSpline
f2 = CubicSpline(x, y, bc_type='natural')
x2 = np.linspace(-5, 5, 100)
y2 = f2(x2)
plt.plot(x2, y2,'g')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()

#Show all 3 graphs together
#Interlaying the graphs show the increase in accuracy when plotting.
plt.figure
plt.plot(x, y, 'o')
plt.plot(x1, y1,'r')
plt.plot(x2, y2, 'g')
plt.show()
```
